### Good Coding Practice: docstrings

In Python, docstrings are a way to document code, making it easier to understand and use. Docstrings are strings enclosed in triple quotes (single or double) that provide information about a module, class, method, or function. Properly documented code is not only helpful to others but also to your future self when revisiting your code.

1. Module-Level Docstrings

A module-level docstring provides an overview of the entire module.
It's placed at the top of the Python file.
It typically includes information about the module's purpose, contents, and usage.

```python
"""
This module contains utility functions for mathematical operations.
"""
```

2. Function and Method Docstrings

Function and method docstrings describe the purpose, parameters, and return value of a function or method.

They are placed immediately below the function or method definition.

```python
def add(a, b):
    """
    Adds two numbers.

    Args:
        a (int or float): The first number.
        b (int or float): The second number.

    Returns:
        int or float: The sum of the two input numbers.
    """
    return a + b
```

3. Class Docstrings

Class docstrings provide information about the purpose and usage of a class.

They are placed immediately below the class definition.

```python
class Rectangle:
    """
    Represents a rectangle with width and height.

    Attributes:
        width (float): The width of the rectangle.
        height (float): The height of the rectangle.
    """
    def __init__(self, width, height):
        self.width = width
        self.height = height
```

5. NumPy Operations


```python
import numpy as np
# Element-wise operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
result = arr1 + arr2  # more examples include subtraction, multiplication, division, etc..
print(result)


# Broadcasting is a powerful feature in NumPy that allows you to perform operations on arrays with different shapes.
result = result + 3   # we can also broadcast vectors to matrices etc..
print(result)


# Universal functions are functions that operate element-wise on arrays and provide fast numerical computations.
arr = np.array([1, 2, 3])
sqrt_arr = np.sqrt(arr)  # [1.0, 1.41421356, 1.73205081]
exp_arr = np.exp(arr)    # [2.71828183, 7.3890561, 20.08553692]
sin_arr = np.sin(arr)    # [0.84147098, 0.90929743, 0.14112001]


# NumPy provides functions for aggregating data in arrays, such as finding minimum, maximum, sum, mean, and more.
arr = np.array([1, 2, 3, 4, 5])

min_val = np.min(arr)     # 1
max_val = np.max(arr)     # 5
sum_val = np.sum(arr)     # 15
mean_val = np.mean(arr)   # 3.0
```

```python
# NumPy allows you to sort and search for elements in arrays.
arr = np.array([3, 1, 2, 4, 5])
# Sorting
sorted_arr = np.sort(arr)  # [1, 2, 3, 4, 5]
# Searching
index = np.where(arr == 4)  # Returns the index where 4 is found


# NumPy provides various functions for manipulating arrays, including reshaping, stacking, and concatenating.
arr = np.array([[1, 2], [3, 4]])
# Reshaping the array
reshaped_arr = arr.reshape((4, 1))  # [[1], [2], [3], [4]]
# Stacking arrays vertically
stacked_arr = np.vstack((arr, arr))  # [[1, 2], [3, 4], [1, 2], [3, 4]]
# Concatenating arrays
concatenated_arr = np.concatenate((arr, arr), axis=1)  # [[1, 2, 1, 2], [3, 4, 3, 4]]
```

```python
# Array Indexing and Slicing
arr = np.array([[1, 2, 3], [3, 4, 5]])
A1 = arr[0]  # [1, 2, 3]
A11 = arr[0][0]  # 1
subarray = arr[0:1, 0:2]  # for the first dimension, we only take the 0th element, and for the second dimension, we take the first 2 elements
print(subarray)
subarray = arr[:, 0:2]
# print(subarray)


# Boolean Indexing
arr = np.array([1, 2, 3, 4, 5, 3, 2, 1, 2, 3, 1])
mask = arr > 2  # Create a boolean mask, value 1 when the corresponding index in the array is >2 otherwise 0
filtered_array = arr[mask]  # Get elements > 2
print(mask)
print(filtered_array)
```

### Introduction to SciPy

1. What is SciPy?

SciPy is an open-source library for mathematics, science, and engineering.

It builds on NumPy and provides additional functionality for optimization, integration, interpolation, linear algebra, and more.

2. Importing SciPy

Import SciPy into your Python script or Jupyter Notebook:

`import scipy as sp`

3. Key Submodules of SciPy

*    sp.optimize: Optimization routines.
*    sp.integrate: Integration and ODE solving.
*    sp.interpolate: Interpolation and curve fitting.
*    sp.linalg: Linear algebra operations.
*    sp.fftpack: Fast Fourier Transform (FFT).
*    sp.signal: Signal processing tools.
*    sp.stats: Statistical functions.

**Examples:**

```python
# Optimization with SciPy
from scipy.optimize import minimize

# Define an objective function
def objective(x):
    return x**2 + 4*x + 4

# Minimize the function
result = minimize(objective, x0=0)
print(result)
```

```python
# Integration with SciPy
from scipy.integrate import quad

# Define the function to integrate
def integrand(x):
    return x**2

# Perform integration
result, error = quad(integrand, 0, 1)  # Integrate func from a to b (possibly infinite interval) using a technique from the Fortran library QUADPACK.
print(result, error)
```

```python
# Interpolation with SciPy. Interpolate data points to create a piecewise linear continuous function.

from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Define data points
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Create an interpolation function
f = interp1d(x, y, kind='linear')



######### visualize the plot, we will cover ploting in a later lecture #########
# Generate x values for the plot
x_new = np.linspace(0, 4, 100)  # 100 points between 0 and 4
# Evaluate the interpolation function at x_new
y_new = f(x_new)
# Plot the original data points and the interpolation
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Original Data', color='red')
plt.plot(x_new, y_new, label='Interpolation', linestyle='--', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Interpolation')
plt.grid(True)
plt.show()
```

```python
# Linear Algebra with SciPy
import scipy as sp
from scipy.linalg import inv, det, eig

# Define a matrix
A = np.array([[1, 2], [3, 4]])
# Matrix inversion
A_inv = inv(A)
# Determinant calculation
A_det = det(A)
# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = eig(A)
```

```python
# Signal Processing with SciPy
from scipy import signal
import numpy as np

# Generate a simple signal
t = np.linspace(0, 10, 1000, endpoint=False)
signal_data = np.sin(2 * sp.pi * 1 * t) + 0.5 * np.random.normal(size=len(t))

# Apply a low-pass Butterworth filter
b, a = signal.butter(4, 0.1)
filtered_signal = signal.filtfilt(b, a, signal_data)



# Plot the original signal and the filtered signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, signal_data, label='Original Signal', color='blue')
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(t, filtered_signal, label='Filtered Signal', color='red')
plt.title('Filtered Signal (Low-pass Butterworth Filter)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# **Lecture 4 In-Class Exercise** (OOP)



[This is a Leetcode problem.](https://leetcode.com/problems/same-tree/)

Design a parking system for a parking lot. The parking lot has three kinds of parking spaces: big, medium, and small, with a fixed number of slots for each size.

Implement the `Parkingsystem` class:

- `ParkingSystem(int big, int medium, int small)` Initializes object of the ParkingSystem class. The number of slots for each parking space are given as part of the constructor.
- `bool addCar (int carType)` Checks whether there is a parking space of `carType` for the car that wants to get into the parking lot. `carType` can be of three kinds: big, medium, or small, which are represented by 1,2, and 3 respectively. A car can only park in a parking space of its `carType`. If there is no space available, return `false`, else park the car in that size space and return true.


---
**Example 1:**

**Input**

["ParkingSystem", "addCar", "addCar", "addCar", "addCar"]

[[1, 1, 0], [1], [2], [3], [1]]

**Output**

[null, true, true, false, false]

**Explanation**

ParkingSystem parkingSystem = new ParkingSystem(1, 1, 0);

parkingSystem.addCar(1); // return true because there is 1 available slot for a big car

parkingSystem.addCar(2); // return true because there is 1 available slot for a medium car

parkingSystem.addCar(3); // return false because there is no available slot for a small car

parkingSystem.addCar(1); // return false because there is no available slot for a big car. It is already occupied.

```python
class ParkingSystem:

    def __init__(self, big: int, medium: int, small: int):
        self.capacity = [big, medium, small]

    def addCar(self, carType: int) -> bool:
        if self.capacity[carType -1] >0:
            self.capacity[carType -1] -= 1
            return True
        else:
            return False


parkingSystem = ParkingSystem(1, 1, 0)
print(parkingSystem.addCar(1))
print(parkingSystem.addCar(2))
print(parkingSystem.addCar(3))
print(parkingSystem.addCar(1))
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSameTree(self, p, q):
        if not p and not q:
            return True
        if not p or not q or p.val != q.val:
            return False
        return self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)
```
